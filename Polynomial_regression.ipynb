Remove Ouliers


```python
def  wisker(col):
    q1,q3 =np.percentile(col,[25,75])
    iqr = q3-q1
    lw = q1 - 1.5*iqr
    uw = q3 + 1.5*iqr
    return lw,uw
```

```python
for  i in  ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'black', 'lstat','medv']:
    lw , uw = wisker(df[i])
    df[i] = np.where(df[i]<lw ,lw, df[i])
    df[i] = np.where(df[i]>uw ,uw, df[i])
```

Model building

```python
#without scaling
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train ,test split
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('medv' , axis = 1)
y = df['medv']
```

```python
X
```

```python
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=23)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

X_train = pd.DataFrame(scalar.fit_transform(X_train))
X_test = pd.DataFrame(scalar.fit_transform(X_test))
```

applying other model to improve accuracy

```python
from sklearn.preprocessing import PolynomialFeatures
poly =PolynomialFeatures(degree=2)
```

```python
X_train =  poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
X_train
```

```python
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score ,mean_squared_error
r2_score(y_test,y_pred)
```

```python
y_pred1 = model.predict(X_train)
r2_score(y_train,y_pred1)
```

ridge and lasso regression


```python
from sklearn.linear_model import Ridge , Lasso
#Ridge regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train,y_train)
r_pred  = ridge.predict(X_test)
r = r2_score(y_test,r_pred)
print(r)
#best accuracy occurring at ridge 
```

```python
#Lasso regression
lasso = Lasso(alpha=0.1)
lasso.fit(X_train,y_train)
l_pred  = lasso.predict(X_test)
l = r2_score(y_test,l_pred)
print(l)
```
