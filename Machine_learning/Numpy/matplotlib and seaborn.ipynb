# Seaborn Library

```python
import seaborn as sns
df1 = sns.load_dataset("tips")
df1
```

```python
df1=df1.drop(columns = ['sex','smoker','day','time'])
df1
```

```python
df1.corr()
```

```python
sns.heatmap(df1.corr())
```

# Joinplot

```python
sns.jointplot(x = 'tip' , y = 'total_bill', data = df1 , kind = "hex")
# sns.jointplot(x = 'tip' , y = 'total_bill', data = df1)
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill', data = df1 , kind = "reg")
#distribution and liear regression is also visible
```
